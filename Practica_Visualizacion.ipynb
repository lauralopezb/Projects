Siga las instrucciones para recrear los plots usando esta información:

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

**Importe `matplotlib.pyplot` como `plt` y configure `%matplotlib` en línea si está usando el jupyter notebook.**

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Ejercicio 1

Siga con estos pasos:
* Crea un objeto figura llamado fig usando `plt.figure()`
* Use `add_axes` para agregar un eje al canvas de la figura en `[0,0,1,1]`. Llamar a este nuevo eje ax.
* Grafica (x, y) en esos ejes y establece las etiquetas y títulos para que coincidan con el siguiente diagrama:

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y,'b')
ax.set_xlabel('x') 
ax.set_ylabel('y')
ax.set_title('title');
```
